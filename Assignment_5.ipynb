# Assignment 5: For Loops, Range, and Code Blocks

#### Task 1: Print all even numbers from 1 to 10.

````python
for i in range(2,11,2):
  print(i)
````

#### Task 2: Calculate the sum of numbers from 1 to 100.


````python
total = sum(range(1,101))
print("The sum of numbers from 1 to 100 is: ",total)
````

#### Task 3: Loop through a fixed list of fruits and print each fruit.

````python
fruits = ["apple","banana","cherry","date"]
for fruit in fruits:
  print(fruit)
````
