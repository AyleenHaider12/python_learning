#### Task 1: Create a function that takes two fixed numbers as input and returns their sum.

````python
def sum_two_numbers(num1,num2):
  return num1 + num2
num1= 8
num2= 9
result= sum_two_numbers(num1, num2)
print(f"The sum is: {result}")
````

#### Task 2: Create a function that takes two strings and returns the longer one.

````python
def find_longer_string(str1, str2):
  if len(str1)> len(str2):
    return str1
  elif len(str2)> len(str1):
    return str2
  else:
    return "Both strings are of equal length"
string1 = "Hello"
string2 = "Programming"
result = find_longer_string(string1, string2)
print(f"The longer string is: {result}")
````
