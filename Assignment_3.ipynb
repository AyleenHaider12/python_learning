#### Task 1: Determine if a number is even or odd with a fixed number


````python
number= 4
if number % 2== 0:
  print(f"{number} is even.")
else:
  print(f"{number} is odd.")
````

#### Task 3: Check if a fixed number is positive, negative, or zero.


````python
num= -5
if num > 0:
  print(f"{num} is positive.")
elif num < 0:
  print(f"{num} is negative.")
else:
  print(f"{num} is zero.")
````
