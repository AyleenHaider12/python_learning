# Assignment 4: Randomization and Python Lists

#### Task 1: Generate a random number between 1 and 10 with a fixed seed.


````python
import random
random.seed(42)
random_number = random.randint(1,10)
print(f"Random number: {random_number}")
````

#### Task 2: Create a list of colors and randomly pick one with a fixed seed.


````python
colors = ["blue","green","red","yellow","purple"]
random.seed(42)
chosen_color = random.choice(colors)
print(f"The chosen color is: {chosen_color}")
````

#### Task 3: Shuffle a list of numbers and print it with a fixed seed.


````python
numbers = [1,2,3,4,5]
random.seed(42)
random.shuffle(numbers)
print(f"Shuffled list: {numbers}")
````

###  Sample Outputs


```
Random number: 10
The chosen color is: blue
Shuffled list: [2, 5, 3, 1, 4]
```
