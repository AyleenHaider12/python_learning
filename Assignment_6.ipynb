#### Task 1: Create a function that counts the number of vowels in a given fixed string.





````python
myString=input("Enter string:")
vowels=0
for i in myString:
      if(i=='a' or i=='e' or i=='i' or i=='o' or i=='u' or i=='A' or i=='E' or i=='I' or i=='O' or i=='U'):
            vowels=vowels+1
print("Number of vowels are:")
print(vowels)
````

#### Task 2: Create a while loop that calculates the sum of all numbers from 1 to 50.

````python
sum_of_numbers = 0
i = 1
while i <= 50:
  sum_of_numbers += i
  i += 1
print(f"The sum of numbers from 1 to 50 is: {sum_of_numbers}")
````
