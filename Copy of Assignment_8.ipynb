````python
student_scores = {'Alice': 85, 'Bob': 90, 'Charlie': 78 }
print(student_scores)
````

#### Task 2: Create a nested dictionary with student names and their details.

````python
student_details  = {
    'Alice': {'age': 20, 'major': 'Computer Science'},
    'Bob': {'age': 22, 'major': 'Physics'}
}
print(student_details)
````


#### Task 3: Create a fixed list of dictionaries.

````python
student_list = [{'name': 'Alice', 'score': 85}, {'name': 'Bob', 'score':90}]
print(student_list)
````
